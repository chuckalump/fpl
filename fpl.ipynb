```python
import pandas as pd
import numpy as np
import json
#statsURL="https://fantasy.premierleague.com/api/bootstrap-static/"

#dfStats = pd.read_json("data/ft.json")
file = open("stats_2018_19.json", encoding="utf8")

x  = file.read()
file.close()
dat=json.loads(x)

######################
dfevents=pd.io.json.json_normalize(dat['events'])
dfsettings =pd.io.json.json_normalize(dat['game_settings'])
dfphases =pd.io.json.json_normalize(dat['phases'])
dfteams =pd.io.json.json_normalize(dat['teams'])
#dftotal_players =pd.io.json.json_normalize(dat['total_players'])
dfelements =pd.io.json.json_normalize(dat['elements'])
dfelement_stats =pd.io.json.json_normalize(dat['element_stats'])
dfelement_types =pd.io.json.json_normalize(dat['element_types'])
```

```python
dfpl=dfelements.merge(dfelement_types,left_on='element_type',right_on ='id' ).merge(dfteams, left_on='team', right_on='id') 
dfpl.keys()
```

```python
dfpl.sort_values(['singular_name_short','minutes','total_points'], inplace=True, ascending=False)
dfpl_list=dfpl[dfpl['singular_name_short'].str.contains('DEF')].loc[:,['short_name','first_name','second_name','singular_name_short','total_points','points_per_game', 'bonus','now_cost','minutes', 'goals_scored','assists', 'clean_sheets','saves', 'bps', 'influence', 'creativity', 'threat', 'goals_conceded', 'own_goals','penalties_saved', 'penalties_missed', 'yellow_cards', 'red_cards','chance_of_playing_this_round']]

my_gkp=['de Gea']
my_def =['Aké','van Dijk',  'Duffy','Chilwell']
my_mid =['Zaha','Antonio','Milivojevic',]
my_sub=['Robertson','Redmond','Pereira Gomes','Foster']
my_fwd=['Wilson','Deulofeu','Jiménez']
         
def color(val):
    if val in my_gkp:
        color = 'yellow'
    elif val in my_def:
        color = 'green'
    elif val in my_mid:
        color = 'blue'
    elif val in my_fwd:
        color = 'red'
    elif val in my_sub:
        color = 'orange'
    else :
        color='white'
    return 'background-color: %s' % color

dfpl_list.style.applymap(color, subset=['second_name'])
```
